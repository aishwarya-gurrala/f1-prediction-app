```python
import os, glob
import pandas as pd

# ---- CONFIG ----
DATA_DIR = r"C:\Users\aishw\OneDrive\Desktop\F1predictoncap\Data"
OUT_FILE = os.path.join(DATA_DIR, "f1_data.csv")

def _load(name):
    """Load .xlsx or .csv by base name from DATA_DIR, with clear error if missing."""
    for ext in (".xlsx", ".csv"):
        path = os.path.join(DATA_DIR, name + ext)
        if os.path.exists(path):
            if ext == ".xlsx":
                return pd.read_excel(path)
            else:
                return pd.read_csv(path)
    raise FileNotFoundError(f"Could not find {name}.xlsx or {name}.csv in {DATA_DIR}")

print("Loading tables from:", DATA_DIR)

# ---- LOAD ----
races                = _load("races")
results              = _load("results")
drivers              = _load("drivers")
constructors         = _load("constructors")
qualifying           = _load("qualifying")
driver_standings     = _load("driver_standings")
constructor_standings= _load("constructor_standings")

# Lowercase columns for easier handling (keeps original frames intact)
def lower_cols(df):
    df = df.copy()
    df.columns = [c.lower() for c in df.columns]
    return df

races, results = map(lower_cols, [races, results])
drivers, constructors = map(lower_cols, [drivers, constructors])
qualifying = lower_cols(qualifying)
driver_standings = lower_cols(driver_standings)
constructor_standings = lower_cols(constructor_standings)

# ---- REQUIRED KEYS: standardize typical Ergast names ----
# results keys
if "raceid" not in results.columns or "driverid" not in results.columns or "constructorid" not in results.columns:
    raise KeyError("results must contain columns: raceId, driverId, constructorId")

# races minimal fields
race_keep = []
for cand in ["raceid","year","round","circuitid","name"]:
    if cand in races.columns: race_keep.append(cand)
if not {"raceid","year","round","name"}.issubset(set(race_keep)):
    raise KeyError("races must have at least: raceId, year, round, name")

# qualifying: use qualifying position as 'qual_position' (may be 'position' or 'positionorder')
q_pos_col = "position" if "position" in qualifying.columns else ("positionorder" if "positionorder" in qualifying.columns else None)
if q_pos_col is None:
    # Some exports call it 'qualifyingposition'
    q_pos_col = "qualifyingposition" if "qualifyingposition" in qualifying.columns else None
if q_pos_col is None:
    # If not present, we'll proceed without qualifying position
    print("⚠️  No qualifying position column found; skipping qualifying merge.")
    qualifying = None

# results finishing order column (target helper): usually 'positionorder'
res_pos_order = "positionorder" if "positionorder" in results.columns else None
if res_pos_order is None:
    # sometimes called 'positionOrder' or fall back to numeric 'position'
    if "position" in results.columns:
        res_pos_order = "position"
    else:
        raise KeyError("results needs a finishing order column: positionOrder/position")

# grid (start position) often already exists in results; if not, we can derive from qualifying
has_grid_in_results = "grid" in results.columns

# ---- MERGE CORE ----
df = results.merge(races[race_keep], on="raceid", how="left", validate="many_to_one")

# drivers core info
drv_keep = [c for c in ["driverid","forename","surname","nationality","code"] if c in drivers.columns]
df = df.merge(drivers[drv_keep], on="driverid", how="left", validate="many_to_one")

# constructors core info
team_name_col = "name" if "name" in constructors.columns else ("constructorref" if "constructorref" in constructors.columns else None)
cons_keep = ["constructorid"]
if team_name_col: cons_keep += [team_name_col]
df = df.merge(constructors[cons_keep], on="constructorid", how="left", validate="many_to_one", suffixes=("", "_constructor"))

# qualifying merge (optional)
if qualifying is not None:
    q_keep = ["raceid","driverid", q_pos_col]
    df = df.merge(qualifying[q_keep].rename(columns={q_pos_col: "qual_position"}), on=["raceid","driverid"], how="left", validate="many_to_one")

# Grid: prefer results.grid; else use qualified position as proxy
if has_grid_in_results:
    grid_series = df["grid"]
else:
    grid_series = df.get("qual_position")  # may be NaN if qualifying missing
df["grid"] = grid_series

# ---- STANDINGS to get PRE-RACE cumulative points (leakage-safe via shift) ----
# driver standings
if not {"raceid","driverid","points"}.issubset(driver_standings.columns):
    raise KeyError("driver_standings must have raceId, driverId, points")
ds = driver_standings.sort_values(["driverid","raceid"]).copy()
ds["driver_points_to_date"] = ds.groupby("driverid")["points"].shift(1).fillna(0)
df = df.merge(ds[["raceid","driverid","driver_points_to_date"]], on=["raceid","driverid"], how="left", validate="many_to_one")

# constructor standings
if not {"raceid","constructorid","points"}.issubset(constructor_standings.columns):
    raise KeyError("constructor_standings must have raceId, constructorId, points")
cs = constructor_standings.sort_values(["constructorid","raceid"]).copy()
cs["constructor_points_to_date"] = cs.groupby("constructorid")["points"].shift(1).fillna(0)
df = df.merge(cs[["raceid","constructorid","constructor_points_to_date"]], on=["raceid","constructorid"], how="left", validate="many_to_one")

# ---- TARGET: win flag from finishing order ----
df["win"] = (pd.to_numeric(df[res_pos_order], errors="coerce") == 1).astype(int)

# team name column standardization
if team_name_col:
    df.rename(columns={team_name_col: "team"}, inplace=True)
else:
    df["team"] = None

# ---- SELECT MODEL-FRIENDLY COLUMNS ----
keep_cols = []
for col in ["year","round","grid","qual_position","driver_points_to_date","constructor_points_to_date","win"]:
    if col in df.columns:
        keep_cols.append(col)

# include driver/team text fields (helpful for EDA; you can drop before training)
for extra in ["forename","surname","nationality","team"]:
    if extra in df.columns: keep_cols.append(extra)

# results may also have 'points' (POST-race). DO NOT keep it as a feature.
final_df = df[keep_cols].copy()

# Save
final_df.to_csv(OUT_FILE, index=False)
print(f"✅ Saved: {OUT_FILE}")
print("Shape:", final_df.shape)
display(final_df.head(5))
```

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, brier_score_loss

df = pd.read_csv(r"C:\Users\aishw\OneDrive\Desktop\F1predictoncap\Data\f1_data.csv")
df.head()
```

```python
# Drop non-numeric fields (driver names etc.)
X = df.select_dtypes(include=["number"]).drop(columns=["win"], errors="ignore")
y = df["win"]

# Sort chronologically to avoid leakage
df = df.sort_values(["year", "round"])
```

```python
train = df["year"] < df["year"].max()   # Train on all past seasons
test  = df["year"] == df["year"].max()  # Test on the most recent season

X_train, y_train = X[train], y[train]
X_test,  y_test  = X[test], y[test]
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, brier_score_loss

model = XGBClassifier(
    random_state=42,
    n_estimators=400,
    max_depth=4,
    learning_rate=0.05
)

model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]
y_pred = (y_prob >= 0.5).astype(int)

print("Accuracy:", round(accuracy_score(y_test, y_pred), 3))
print("ROC-AUC:", round(roc_auc_score(y_test, y_prob), 3))
print("Brier Score:", round(brier_score_loss(y_test, y_prob), 3))
```

```python
importance = model.get_booster().get_score(importance_type='gain')
for k, v in sorted(importance.items(), key=lambda x: x[1], reverse=True):
    print(f"{k}: {v:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.hist(y_prob, bins=20, color='skyblue', edgecolor='black')
plt.title("Predicted Win Probabilities Distribution")
plt.xlabel("Win Probability")
plt.ylabel("Number of Drivers")
plt.show()
```
