```python
import os
import numpy as np
import pandas as pd

from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, roc_auc_score, average_precision_score,
    brier_score_loss, precision_score, recall_score, f1_score,
    roc_curve, auc
)

import matplotlib.pyplot as plt

CSV_PATH = r"C:\Users\aishw\OneDrive\Desktop\F1predictoncap\Data\f1_data.csv"

LEAKY_COLS = {"points", "position", "positionorder"}

REQUIRED_FEATURES = [
    "grid",
    "qual_position",
    "driver_points_to_date",
    "constructor_points_to_date",
    "year",
    "round",
]

RENAME_MAP = {
    # If your CSV used different names, map here, e.g.:
    # "qualifying_position": "qual_position",
    # "driver_pts_to_date": "driver_points_to_date",
    # "constructor_pts_to_date": "constructor_points_to_date",
}
```

```python
import pandas as pd
import numpy as np

def assert_no_leakage(columns):
    lower = {c.lower() for c in columns}
    bad = lower & LEAKY_COLS
    if bad:
        raise ValueError(f"Leaky columns detected: {bad}")

# 1. Load data
df = pd.read_csv(CSV_PATH)
if RENAME_MAP:
    df = df.rename(columns=RENAME_MAP)

# 2. Basic checks
if "win" not in df.columns:
    raise KeyError("Target column 'win' not found in CSV.")
if not {"year", "round"}.issubset(df.columns):
    raise KeyError("'year' and 'round' must be in CSV.")

numeric = df.select_dtypes(include=["number"])
assert_no_leakage(numeric.columns)

# 3. Build X, y based on REQUIRED_FEATURES
missing = [c for c in REQUIRED_FEATURES if c not in numeric.columns]
if missing:
    raise KeyError(f"Missing required numeric columns: {missing}")

X = numeric[REQUIRED_FEATURES].copy()
y = df["win"].astype(int)

# 4. DROP rows with any NaN in the feature set (for all models)
mask_valid = ~X.isna().any(axis=1)

df = df[mask_valid].reset_index(drop=True)
X = X[mask_valid].reset_index(drop=True)
y = y[mask_valid].reset_index(drop=True)

print("After dropping NaNs:", X.shape)

# 5. Time-based split: last season as test
last_season = int(df["year"].max())
train_mask = df["year"] < last_season
test_mask  = df["year"] == last_season

X_train, X_test = X[train_mask], X[test_mask]
y_train, y_test = y[train_mask], y[test_mask]

last_season, X_train.shape, X_test.shape
```

```python
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    prob = model.predict_proba(X_test)[:, 1]
    pred = (prob >= 0.5).astype(int)

    metrics = {
        "model": name,
        "accuracy": accuracy_score(y_test, pred),
        "roc_auc": roc_auc_score(y_test, prob),
        "pr_auc":  average_precision_score(y_test, prob),
        "brier":   brier_score_loss(y_test, prob),
        "precision": precision_score(y_test, pred, zero_division=0),
        "recall":    recall_score(y_test, pred, zero_division=0),
        "f1":        f1_score(y_test, pred, zero_division=0),
        "prob":      prob,  # keep probabilities for ROC curves
    }
    return metrics
```

```python
results = []

# 1) Logistic Regression (baseline)
log_reg = LogisticRegression(max_iter=1000, class_weight="balanced")
results.append(evaluate_model("Logistic Regression", log_reg, X_train, y_train, X_test, y_test))

# 2) Random Forest
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    class_weight="balanced_subsample",
)
results.append(evaluate_model("Random Forest", rf, X_train, y_train, X_test, y_test))

# 3) XGBoost (same config as in your app.py)
xgb = XGBClassifier(
    random_state=42,
    n_estimators=400,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.9,
    n_jobs=-1,
)
results.append(evaluate_model("XGBoost", xgb, X_train, y_train, X_test, y_test))

# Make a clean table without 'prob'
df_results = pd.DataFrame([
    {k: v for k, v in r.items() if k != "prob"} for r in results
])
df_results
```

```python
plt.figure(figsize=(6, 4))

for res in results:
    name = res["model"]
    prob = res["prob"]
    fpr, tpr, _ = roc_curve(y_test, prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f"{name} (AUC = {roc_auc:.3f})")

plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random baseline")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curves on Held-Out Season {last_season}")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()

plt.savefig("roc_model_comparison.png", dpi=200)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# probability scores for class 1
xgb_prob = results[2]["prob"]   # XGBoost probability scores

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, xgb_prob)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(7,6))
plt.plot(fpr, tpr, color='orange', lw=3,
         label=f"XGBoost ROC curve (AUC = {roc_auc:.3f})")

plt.plot([0,1], [0,1], linestyle='--', color='navy', lw=2)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve  XGBoost F1 Win Prediction")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```
